```python
import math as mt

def combinations(n, k):
    return int(mt.factorial(n)/(mt.factorial(k)*mt.factorial(n-k)))

def print_combinations(n, k):
    return f"{n}!/({k}!*({n}-{k})!)"

print("1. Из колоды в 52 карты извлекаются случайным образом 4 карты.\na) Найти вероятность того, что все карты – крести;\nб) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.")
cards_count = 52
n = 4
m = int(cards_count/4)
q = combinations(m, n)
print(f"\nОтвет.\n\nа) Число карт крестовой масти составляет 1/4 от числа карт в колоде и равно m = {m} ({cards_count}/4).")
print(f"{n} произвольных карты из {m} карт крестовой масти без учёта порядка можно извлечь {print_combinations(m, n)} способами,")
print(f"что составляет {q} способов.")
p = combinations(cards_count, n)
print(f"{n} произвольных карты из колоды в 52 карты без учёта порядка можно извлечь {print_combinations(cards_count, n)} способами,")
print(f"что составляет {p} способов (число всех равновозможных исходов).")
print(f"Вероятность того, что все карты из {n}-х – крести, составит P(A) = {q}/{p} = {q/p:.6f}.")
print(f"\nб) Среди вытянутых {n} карт в идеале должно находиться от 1 до 4 тузов. Тогда число вариантов вытянуть:")

ans = []
for i in range(1, n + 1):
    if i == 1:
        s = "туз "
    else:
        s = "туза"
    if i == 4:
        term_s = "."
    else:
        term_s = ";"
    ans.append(combinations(n, i) * combinations(cards_count - n, n - i))
    print(f"   - {i} {s} - {print_combinations(n, i)} * {print_combinations(cards_count - n, n - i)} = {ans[i - 1]}{term_s}")
m = sum(ans)
print(f"Суммарно {ans[0]} + {ans[1]} + {ans[2]} + {ans[3]} = {m}.")
print(f"Соответственно, искомая вероятность P(A) = {m}/{p} = {m/p:.6f}.")
```

```python
print("2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9.\nКод содержит три цифры, которые нужно нажать одновременно.\nКакова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?")
print("\nОтвет. Пусть событие А - успешное открывание двери подъезда с кодовым замком с вероятностью P(A) = m/n.")
m = 1
n = combinations(10, 3)
print(f"Первая попытка не предполагает учёта предыдущих нажатий, тогда число исходов, благоприятствующих наступлению события А - m = {m}, или {print_combinations(3, 0)}.")
print("С учётом необходимости одновременного нажатия 3-х цифр, т.е. порядок нажатия не важен, число всех равновозможных исходов")
print(f"опыта, в котором событие А может появиться, составит (10*9*8)/(3!), или {print_combinations(10, 3)} = {n}.")
print(f"Соответственно, вероятность открывания двери подъезда с первой попытки P(A) = {m}/{n} = {m/n:.6f}.")
```

```python
print("3. В ящике имеется 15 деталей, из которых 9 окрашены.\nРабочий случайным образом извлекает 3 детали.\nКакова вероятность того, что все извлеченные детали окрашены?")
n = combinations(9, 3)
m = combinations(15, 3)
print(f"\nОтвет. Число благоприятствующих исходов составит {print_combinations(9, 3)} = {n}, равновозможных исходов - {print_combinations(15, 3)} = {m}.")
print(f"Соответственно, искомая вероятность P(A) = {n}/{m} = {n/m:.6f}.")
```

```python
print("4. В лотерее 100 билетов. Из них 2 выигрышных.\nКакова вероятность того, что 2 приобретенных билета окажутся выигрышными?")
k = 2
n = 100
m = combinations(n, k)
print(f"\nОтвет. Приобретение {k} билетов из {n} в произвольной последовательности осуществляется {print_combinations(n, k)} = {m} способами.")
print(f"Соответственно, вероятность приобретения {k} выигрышных билетов P(A) = 1/{m} = {1/m:.6f}.")
```
