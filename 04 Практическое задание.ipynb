```python
from scipy.linalg import lu

def LU_decision(a, b):
    res = [el for el in b]
    i_max = len(a)
    j_max = len(a[0])
    l, u = get_lu_matrix(a)
    
    for i in range (i_max):
        s = b[i]
        for j in range (i + 1):
            if i == j:
                res[j] = s / l[i][j]
            else:
                s -= (l[i][j] * res[j])

    for i in range (i_max - 1, -1, -1):
        s = res[i]
        for j in range (j_max - 1, i - 1, -1):
            if i == j:
                res[j] = s / u[i][j]
            else:
                s -= (u[i][j] * res[j])
    return res

def get_lu_matrix(a):
    i_max = len(a)
    j_max = len(a[0])
    l_m = np.zeros(shape=(i_max, j_max))
    u_m = np.zeros(shape=(i_max, j_max))
    for i in range(i_max):
        for j in range(j_max):
            u_m[i][j] = a[i][j]
    
    for j in range(j_max):
        for i in range(j, i_max):
            if j == i:
                l_m[i][j] = 1
            elif i > j:
                coeff = u_m[i][j] / u_m[j][j]
                l_m[i][j] = coeff
                for k in range(j, j_max):
                    u_m[i][k] -= (coeff * u_m[j][k])
    return l_m, u_m


print('Задания 2 и 5 (с написанием программы для LU-разложения)')
print('--------------------------------------------------------\n')
s = ['a) ', 'б) ']
for i in range(2):
    if i == 0:
        A = np.array([[1, 2, 4], [2, 9, 12], [3, 26, 30]])
    else:
        A = np.array([[1, 1, 2, 4], [2, 5, 8, 9], [3, 18, 29, 18], [4, 22, 53, 33]])
        
    L, U = get_lu_matrix(A)
        
    print(f'{s[i]}Матрица коэффициентов:\n{A}\n')
    print(f'L-матрица:\n{L}\n')
    print(f'U-матрица:\n{U}\n')
    print(f'Проверка - произведение L*U для собственной программы:\n{np.dot(L,U)}\n')

    sp_pl, sp_u = lu(A, permute_l=True)
    print(f'Вариант: PL-матрица для LU-разложения посредством "scipy.linalg.lu":\n{sp_pl}\n')
    print(f'U-матрица для LU-разложения посредством "scipy.linalg.lu":\n{sp_u}\n')
    print(f'Проверка - произведение PL*U для "scipy.linalg.lu":\n{np.dot(sp_pl,sp_u)}\n')
```

```python
def Cholesky_decision(a, b):
    res = [el for el in b]
    i_max = len(a)
    j_max = len(a[0])
    l = np.linalg.cholesky(a)
    u = l.T
    
    for i in range (i_max):
        s = b[i]
        for j in range (i + 1):
            if i == j:
                res[j] = s / l[i][j]
            else:
                s -= (l[i][j] * res[j])

    for i in range (i_max - 1, -1, -1):
        s = res[i]
        for j in range (j_max - 1, i - 1, -1):
            if i == j:
                res[j] = s / u[i][j]
            else:
                s -= (u[i][j] * res[j])
    return res


print('Задания 4 и 5 (решение СЛАУ встроенным разложением Холецкого с собственным нахождением вектора решений)')
print('-------------------------------------------------------------------------------------------------------\n')

A = np.array([[81, -45, 45], [-45, 50, -15], [45, -15, 38]])
B = np.array([531, -460, 193])
print(f'Матрица коэффициентов:\n{A}\n')
print(f'Вектор свободных членов: {B}\n')
L = np.linalg.cholesky(A)
print(f'L-матрица Холецкого:\n{L}\n')
print(f'Проверка L*L.T=A:\n{np.dot(L, L.T)}\n')
X = Cholesky_decision(A, B)
print(f'Вектор решений: {X}')
print(f'Проверка A*X=B: {np.dot(A, X)}')
```
