Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

boston = load_boston()
print(f"Ключи датасета boston: {boston.keys()}")
```

```python
X = pd.DataFrame(boston["data"], columns = boston["feature_names"])
print(X.info())
X.head()
```

```python
Y = pd.DataFrame(boston["target"], columns = ["price"])
print(Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
print(f"506 * 0.3 = {round(506 * 0.3, 0)}\n")
print(Y_test.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
Y_pred = lr.predict(X_test)
print(Y_pred.shape)
```

```python
check_test = pd.DataFrame({"Y_test": Y_test["price"], "Y_pred": Y_pred.flatten()})

check_test.head(10)
```

```python
from sklearn.metrics import r2_score

r2_LR_val = r2_score(Y_test, Y_pred)
print(f"Метрика оценки регрессионных моделей (модель линейной регрессии) R2 = {r2_LR_val}")
```

Вызовите документацию для класса RandomForestRegressor,
найдите информацию об атрибуте feature_importances_.
С помощью этого атрибута найдите сумму всех показателей важности,
установите, какие два признака показывают наибольшую важность.

```python
m = model.feature_importances_
len_m = len(m)
print(f"{m} - массив показателей важности длиной {len_m}.")
print(f"Сумма показателей важности равна {sum(m)}.\n")
highest_imp = np.zeros(4).reshape(2, 2)               # два наибольших показателя важности и их индексы

for i in range(0, len_m):
    if m[i] > highest_imp[0, 0]:
        highest_imp[1, 0] = highest_imp[0, 0]
        highest_imp[1, 1] = highest_imp[0, 1]
        highest_imp[0, 0] = m[i]
        highest_imp[0, 1] = i
    elif m[i] > highest_imp[1, 0]:
        highest_imp[1, 0] = m[i]
        highest_imp[1, 1] = i

print(f"Два наибольших показателя важности - {highest_imp[0, 0]:.8f} (индекс {highest_imp[0, 1]:.0f}) и {highest_imp[1, 0]:.8f} (индекс {highest_imp[1, 1]:.0f}).")
print(f"Им соответствуют признаки '{X.columns[int(highest_imp[0, 1])]}' и '{X.columns[int(highest_imp[1, 1])]}'.")
```

```python
clf.fit(X_train, Y_train)
```

```python
from sklearn.datasets import load_wine

data = load_wine()
print(f"Тип структуры data: {type(data)}")
data_keys = data.keys()
print(f"\nКлючи структуры data: {data_keys}")
print(f"\nКлассы целевой переменной ({len(data.target_names)} шт.): {data.target_names}")
print(f"\nОписание признаков структуры:\n\n{data['DESCR']}")
```

```python
X = pd.DataFrame(data["data"], columns = data["feature_names"])
print(f"Размер: {X.shape}\n")
print(X.info())
```

```python
print(f"Число пропусков данных по столбцам:\n\n{X.isnull().sum(axis=0)}")
```

```python
X = X.assign(target = X['magnesium'] // 50  + 2 * (X['malic_acid'] + X['total_phenols'] + 2.1 * X['hue'] + X['flavanoids'] + X['nonflavanoid_phenols']) // 3)
print(X.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (15, 10))
sns.set(font_scale = 1.4)

X_corr = X.corr()
X_corr = np.round(X_corr, 2)
#X_corr[np.abs(X_corr) < 0.3] = 0

sns.heatmap(X_corr, annot = True, linewidths = .5, cmap = 'coolwarm')

plt.title('Корреляционная матрица')
plt.show()
```

```python
arr = np.array(X_corr)
fields = [str(el) for el in X.columns]
target_ind = fields.index('target')
high_corr = []

for i in range(0, arr.shape[1]):
    if np.abs(arr[target_ind, i]) > 0.5 and i != target_ind:
        high_corr.append(fields[i])

print(high_corr)
X_corr
```

```python
X = X.drop("target", axis = 1)

for el in high_corr:
    s = el + '_2'
    X.loc[:, s] = X[el] ** 2
    
print(X.describe())
```
