```python
print("1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].\n   Найдите её среднее значение и дисперсию.")
a = 200
b = 800
print(f"\nОтвет. Среднее значение M(A) = ({a} + {b})/2 = {(a + b)/2:.0f}, дисперсия D(A) = ({b} - {a})**2/12 = {(b - a)**2/12:.0f}.")
```

```python
import math

print("2. О случайной непрерывной равномерно распределенной величине B известно, что её дисперсия равна 0.2.\n   Можно ли найти правую границу величины B и её среднее значение зная, что левая граница равна 0.5?\n   Если да, найдите её.")
print("\nОтвет. Пусть a = 0.5 - левая граница, b - правая граница. Отсюда b > a.")
a = 0.5
b = math.sqrt(2.4) + a
print("Дисперсия D(B) = 0.2 = (b - a)**2/12. Следовательно, (b - 0.5)**2 = 12 * 0.2 = 2.4.")
print(f"С учётом b > 0.5 правая граница b = sqrt(2.4) + 0.5 = {b:.3f}.")
print(f"Наконец, среднее значение M(B) = ({a:.3f} + {b:.3f})/2 = {(a + b)/2:.3f}.")
```

```python
from scipy.stats import norm

def prob(val):
    return norm.cdf((val - 174) / 8)

print("4. Рост взрослого населения города X имеет нормальное распределение,\n   причём средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.\n   Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:\n   а) больше 182 см; б) больше 190 см;\n   в) от 166 см до 190 см; г) от 166 см до 182 см; д) от 158 см до 190 см;\n   е) не выше 150 см или не ниже 190 см; ё) не выше 150 см или не ниже 198 см; ж) ниже 166 см.")
print(f"\nОтвет. Вероятность роста: а) больше 182 см - {1 - prob(182):.4f}; б) больше 190 см - {1 - prob(190):.4f}; в) от 166 см до 190 см - {prob(190) - prob(166):.4f};")
print(f"г) от 166 см до 182 см - {prob(182) - prob(166):.4f}; д) от 158 см до 190 см - {prob(190) - prob(158):.4f}; е) не выше 150 см или не ниже 190 см - {1 - prob(190) + prob(150):.4f};")
print(f"ё) не выше 150 см или не ниже 198 см - {1 - prob(198) + prob(150):.4f}; ж) ниже 166 см - {prob(166):.4f}.")
```
