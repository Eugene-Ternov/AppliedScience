```python
from math import sqrt
from scipy.stats import norm
from scipy.stats import t as t_cr

print("1. Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16.\n   Найти доверительный интервал для оценки математического ожидания A с надежностью 0.95, если выборочная средняя M = 80,\n   а объем выборки n = 256.")
s = 16    # СКО генеральной совокупности
n = 256   # Объём выборки
m = 80    # Выборочная средняя
z = round(norm.ppf(1 - 0.05/2), 3)
std = s/sqrt(n)
print(f"\nОтвет. Сигма генеральной совокупности известна (s = {s}), следует использовать Z-критерий. С учётом симметричности распределения")
print(f"относительно выборочной средней M = {m} определяем Za/2 для 0.975. Из таблицы Z(0.975) = {z}; стандартная ошибка среднего для")
print(f"выборки объёмом n = {n} составит {s}/sqrt({n}) = {std:.0f}. Наконец, доверительный интервал M +/- (Za/2 * std) = [{m - std*z}; {m + std*z}].")
```

```python
print("2. В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные:\n   6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1. Предполагая, что результаты измерений подчинены нормальному закону\n   распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это\n   значение с доверительной вероятностью 0,95.")

x = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
n = len(x)
m = sum(x) / n
s = sqrt(sum([(el - m)**2 for el in x]) / (n - 1))
t = t_cr.ppf(1 - 0.05/2, n - 1)
half = t * s / sqrt(n)

print("\nОтвет. Сигма генеральной совокупности неизвестна, следует использовать t-критерий (Стьюдента). С учётом симметричности")
print(f"распределения относительно выборочной средней M = {m:.3f} определяем t для a = 0.975 (1 - 0.05/2) и числа степеней свободы v = {n - 1}.")
print(f"Из таблицы t(0.975, {n - 1}) = {t:.3f}; сигма для несмещённой оценки дисперсии s = {s:.3f}. Наконец, доверительный интервал")
print(f"M +/- (t * s/sqrt(n)) = [{m - half:.3f}; {m + half:.3f}].")
```

```python
print("3. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.\n   Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался\n   равным 17.5 мм, а дисперсия известна и равна 4 кв. мм (решить через тестирование гипотезы).")
m0 = 17
m = 17.5
d = 4
n = 100
z = round(norm.ppf(1 - 0.05), 3)
print(f"\nОтвет. Поскольку {m} > {m0}, проверяем гипотезу Н1: m > m0 (H0: m0 = {m0}). Дисперсия известна (D = {d}), используем Z-критерий.")
print(f"Zн = ({m} - {m0})/(sqrt({d})/sqrt({n})) = {(m - m0)/(sqrt(d)/sqrt(n))}. Согласно таблице Z(0.95) = {z}. Zн > Z(0.95) - таким образом, гипотеза Н0\nневерна: истинный средний диаметр шариков в генеральной совокупности превышает 17 мм.")
```

```python
print("4. Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой\n   пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли\n   утверждение продавца, если учитывать, что доверительная вероятность равна 99 % (решить через тестирование гипотезы)?")
cookies = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
m0 = 200
n = len(cookies)
m = sum(cookies)/n
d = sum([(el - m0)**2 for el in cookies])/(n - 1)
t_l = t_cr.ppf(0.005, n - 1)
t_r = t_cr.ppf(1 - 0.005, n - 1)
print(f"\nОтвет. Дисперсия неизвестна, используем двусторонний t-критерий (Стьюдента) для проверки гипотезы H1: m != m0 (H0: m0 = {m0}).")
print(f"Среднее арифметическое выборки m = {m} г, дисперсия для несмещённой оценки D = {d:.1f}. Расчётное значение tн = ({m} - {m0})/")
print(f"/(sqrt({d:.1f})/sqrt({n})) = {(m - m0)/(sqrt(d)/sqrt(n)):.6f}. Для числа степеней свободы v = {n - 1} и доверительной вероятности 99 % следует взять табличные")
print(f"t(0.005, 9) = {t_l:.6f} и t(0.995, 9) = {t_r:.6f}. t(0.005, 9) < tн < t(0.995, 9) - таким образом, верна гипотеза H0, или\nутверждение продавца о том, что средний вес пачки печенья составляет 200 г.")
```
