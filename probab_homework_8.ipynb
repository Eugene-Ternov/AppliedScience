```python
import numpy as np
from scipy import stats

print("1. Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов\n   и штангистов. Даны значения роста в трёх группах случайно выбранных спортсменов:\n   - футболисты - 173, 175, 180, 178, 177, 185, 183, 182;\n   - хоккеисты - 177, 179, 180, 188, 177, 172, 171, 184, 180;\n   - штангисты - 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.")

height_arr = []
height_arr.append(np.array([173, 175, 180, 178, 177, 185, 183, 182]))
height_arr.append(np.array([177, 179, 180, 188, 177, 172, 171, 184, 180]))
height_arr.append(np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170]))

res = stats.f_oneway(height_arr[0], height_arr[1], height_arr[2])
stat = res[0]
pval = res[1]
df1 = len(height_arr)       # 3 группы
df2 = 0                     # общее число спортсменов в 3-х группах
for el in height_arr:
    df2 += len(el)
#df2 = sum(height_arr) - 3
print(f"\nОтвет. Имеем {df2} наблюдений в {df1} группах, расчётный критерий Фишера для однофакторного дисперсионного анализа Fн = {res[0]:.9f}")
print(f"и соответствующее ему P-value = {res[1]:.9f}. Число степеней свободы для определения табличных значений критических точек Fкрит")
df2 -= df1
df1 -= 1
F_005 = 3.385
F_001 = 5.568
print(f"распределения Фишера-Снедекора v1 = {df1} и v2 = {df2}. При указанных v1 и v2 для уровней значимости a = 0.05 и a = 0.01 Fкрит примет")
print(f"значение {F_005} и {F_001} соответственно. Тестируем гипотезу об отсутствии различий среднего роста среди 3 групп спортсменов m1 =")
print(f"= m2 = m3. Для a = 0.05 P-value < a ({res[1]:.6f} < 0.05) или соответственно Fн > Fкрит ({res[0]:.3f} > {F_005}) - средний рост спортсменов")
print(f"в группах различен. Для a = 0.01 P-value > a ({res[1]:.6f} > 0.01) или соответственно Fн < Fкрит ({res[0]:.3f} < {F_001}) - статистическое")
print("отличие среднего роста спортсменов в группах не наблюдается.")
```
