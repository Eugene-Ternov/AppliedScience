```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.stats import t as t_cr

print("1. Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):\n   zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110], ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Найдите\n   ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy. Полученные значения\n   должны быть равны. Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух\n   признаков, а затем с использованием функций из библиотек numpy и pandas.")
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
df = pd.DataFrame({'zp':zp, 'ks': ks}, columns = ['zp', 'ks'])
c = (zp * ks).mean() - zp.mean() * ks.mean()
cv = np.cov(zp, ks, ddof = 0)[0, 1]
r_xy = cv / (zp.std(ddof=0) * ks.std(ddof=0))
print(f"\nОтвет. Ковариация: 1) с помощью элементарных действий c = {c:.5f}; 2) с помощью функции cov = {cv:.5f}.")
print(f"Коэффициент корреляции Пирсона: 1) с помощью ковариации и СКО двух признаков r_xy = {r_xy:.5f}; 2) с помощью функции из библиотеки")
print(f"numpy r_xy = {np.corrcoef(zp, ks)[0, 1]:.5f}; 3) с помощью функции из библиотеки pandas r_xy = {df.corr()['zp']['ks']:.5f}.")
```

```python
print("2. Измерены значения IQ выборки студентов, обучающихся в местных технических вузах:\n   131, 125, 115, 122, 131, 115, 107, 99, 125, 111. Известно, что в генеральной совокупности IQ распределён нормально.\n   Найдите доверительный интервал для математического ожидания с надёжностью 0.95.")
iq = [131, 125, 115, 122, 131, 115, 107, 99, 125, 111]
n = len(iq)
m = np.mean(iq)           # для разноарбузия (виноват: для разнообразия [и навыка использования библиотечных функций])
s = np.std(iq, ddof=1)    # СКО несмещённой оценки
t = t_cr.ppf(1 - 0.05/2, n - 1)
half = t * s / np.sqrt(n)

print("\nОтвет. Поскольку дисперсия генеральной совокупности неизвестна, используем t-критерий. С учётом симметричности распределения")
print(f"относительно выборочной средней M = {m:.3f} определяем t для a = 0.975 (1 - 0.05/2) и числа степеней свободы v = {n - 1}.")
print(f"Из таблицы t(0.975, {n - 1}) = {t:.3f}; сигма для несмещённой оценки дисперсии s = {s:.3f}. Наконец, доверительный интервал")
print(f"M +/- (t * s/sqrt(n)) = [{m - half:.3f}; {m + half:.3f}].")
```

```python
print("3. Известно, что рост футболистов в сборной распределён нормально с дисперсией генеральной совокупности, равной 25 кв.см.\n   Объем выборки равен 27, среднее выборочное составляет 174.2. Найдите доверительный интервал для математического ожидания\n   с надёжностью 0.95.")
d = 25     # СКО генеральной совокупности
m = 174.2  # Выборочная средняя
n = 27     # Объём выборки
z = norm.ppf(1 - 0.05/2)
std_err = np.sqrt(d)/np.sqrt(n)
print("\nОтвет. Поскольку дисперсия генеральной совокупности известна, используем Z-критерий. С учётом симметричности распределения")
print(f"относительно выборочной средней M = {m} определяем Za/2 для 0.975. Из таблицы Z(0.975) = {z:.3f}; стандартная ошибка среднего")
print(f"выборочного для объёма n = {n} равна sqrt({d})/sqrt({n}) = {std_err:.3f}. Наконец, доверительный интервал M +/- (Za/2 * std) = [{m - std_err * z:.3f};\n{m + std_err * z:.3f}].")
```
