```python
import math as mt

salary = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
n = len(salary)
avg_salary = sum(salary) / n
sum_sqr = sum([(el - avg_salary)**2 for el in salary])
offset_D = sum_sqr/n
unbiased_D = sum_sqr/(n - 1)
S = mt.sqrt(offset_D)
S0 = mt.sqrt(unbiased_D)

print("1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84,\n   90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое,\n   среднее квадратичное отклонение, смещённую и несмещённую оценки дисперсий для данной выборки.")
print(f"\nОтвет. Среднее арифметическое: {avg_salary:.2f}; смещённая оценка дисперсии: {offset_D:.3f}; несмещённая оценка дисперсии: {unbiased_D:.3f}.")
print(f"Среднее квадратическое отклонение на основании смещённой оценки дисперсии - {S:.3f}, несмещённой оценки дисперсии - {S0:.3f}.")
```

```python
def combinations(n, k):
    max_val = max(k, n - k)
    min_val = min(k, n - k)
    res = 1
    for i in range(max_val + 1, n + 1):
        res *= i
    return int(res / mt.factorial(min_val))

def print_combinations(n, k):
    return f"{n}!/({k}!*({n}-{k})!)"

print("2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика\n   вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?")
total = combinations(8, 2) * combinations(12, 4)
print(f"\nОтвет. Число всех равновозможных исходов извлечения мячей из обоих ящиков указанным способом равно\n{print_combinations(8, 2)} * {print_combinations(12, 4)} = {total}.")
print("Соответственно, 3 белых мяча можно извлечь как: 1) 2 из 1-го ящика и 1 из 2-го; 2) 1 из 1-го ящика и 2 из 2-го; 2) 3 из 2-го.")
partial = combinations(5, 2) * combinations(5, 1) * combinations(7, 3)
partial += combinations(5, 1) * combinations(3, 1) * combinations(5, 2) * combinations(7, 2)
partial += combinations(3, 2) * combinations(5, 3) * combinations(7, 1)
print(f"Число благоприятствующих исходов равно {print_combinations(5, 2)} * {print_combinations(5, 1)} * {print_combinations(7, 3)} +")
print(f"+ {print_combinations(5, 1)} * {print_combinations(3, 1)} * {print_combinations(5, 2)} * {print_combinations(7, 2)} +")
print(f"+ {print_combinations(3, 2)} * {print_combinations(5, 3)} * {print_combinations(7, 1)} = {partial}.")
print(f"Вероятность того, что 3 мяча белые равна {partial}/{total} = {partial/total:.6f}.")
```

```python
print("3. На соревновании по биатлону один из трёх спортсменов стреляет и попадает в мишень. Вероятность попадания для первого\n   спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведён: a) первым\n   спортсменом; б) вторым спортсменом; в) третьим спортсменом.")
p = [0.9, 0.8, 0.6] 
full_p_goal = sum(p)/3
print("\nОтвет. Вероятность поражения мишени любым из трёх спортсменов равна 1/3. Тогда вероятность того, что мишень будет поражена")
print(f"равна ({p[0]} + {p[1]} + {p[2]})/3 = {full_p_goal:.3f}. Вероятность того, что по мишени выстрелил:")
for i in range(3):
    if i < 2:
        s = ";"
    else:
        s = "."
    print(f"- {i + 1}-й спортсмен - ({p[i]} * 1/3)/{full_p_goal:.3f} = {p[i]/full_p_goal/3:.3f}{s}")
```

```python
print("4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же,\n   сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета\n   B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он\n   учится: a) на факультете A; б) на факультете B; в) на факультете C?")
arr = ((0.8, 0.7, 0.9), (1, 1, 2), ('A', 'B', 'C'))
full_p_goal = 0
sum_arr = sum(arr[1])

for i in range(len(arr)):
    full_p_goal += arr[0][i] * arr[1][i]
full_p_goal /= sum_arr
print("\nОтвет. Пусть на факультеты A и B поступило по X человек, тогда на факультет C - 2 * X человек.")
print(f"Вероятность сдачи 1-й сессии студентами всех факультетов равна X * (0.8 + 0.7 + 2 * 0.9) / (X * (1 + 1 + 2)) и составит {full_p_goal:.4f}.")
print("Вероятность того, что сдавший 1-ю сессию студент учится:")
for i in range(3):
    if i < 2:
        s = ";"
    else:
        s = "."
    print(f"- на факультете {arr[2][i]} - ({arr[0][i]} * {arr[1][i]}/{sum_arr})/{full_p_goal:.4f} = {arr[0][i]*arr[1][i]/full_p_goal/sum_arr:.4f}{s}")
```

```python
import itertools as it

p = [0.1, 0.2, 0.25]
q = [1-el for el in p]
comb = []
s = ""

def get_failure_probability(IsOnlyOne = True):
    global s, comb, p, q
    s_len = len(s)
    comb_len = len(comb)
    res = 0
    for i in range(s_len):
        for j in range (comb_len):
            if comb[j].count(s[i]) == 0:
                mul = 1
                if IsOnlyOne:
                    mul *= p[i]       # Отказ 1 детали
                else:
                    mul *= q[i]       # Безотказная работа 1 детали
                for k in range(2):
                    if IsOnlyOne:
                        mul *= q[int(comb[j][k])]   # Одна отказала - остальные 2 работают
                    else:
                        mul *= p[int(comb[j][k])]   # Одна работает - остальные 2 отказали
                res += mul
                break
    return res

for i in range(len(p)):
    s += str(i)
for el in it.combinations(s, 2):
    comb.append(el)
    
p_one_failed = get_failure_probability()
p_two_failed = get_failure_probability(False)

print("5. Устройство состоит из трёх деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2,\n   для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а) все детали; б) только две детали;\n   в) хотя бы одна деталь; г) от одной до двух деталей?")
print("\nОтвет. Если вероятность отказа i-детали в 1-й месяц работы p(i), то вероятность её безотказной работы q(i) = 1 - p(i).")
print(f"Вероятность отказа в 1-й месяц работы всех деталей равна {p[0]} * {p[1]} * {p[2]} = {p[0] * p[1] * p[2]:.3f}.")
print(f"Для определения вероятностей отказа 1 или 2 деталей с помощью itertools сформируем массив комбинаций индексов '3 по 2':\n{arr}.")
print("Порядок расчёта вероятности отказа 1 или 2 деталей представлен в подпрограмме get_failure_probability.")
print(f"Вероятность отказа в 1-й месяц работы 2 деталей - {p_two_failed:.3f}, 1 детали - {p_one_failed:.3f}, от 1 до 2 деталей - {p_one_failed + p_two_failed:.3f}.")
```
