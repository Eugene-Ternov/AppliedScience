__1.__ Найдите собственные векторы и собственные значения для линейного оператора, заданного матрицей:

$$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$$

```python
import numpy as np

a = np.array([[-1, -6], [2, 6]])
w, v = np.linalg.eig(a)
print(f'Матрица A:\n{a}\n')
print(f'Собственные значения:\n{w}\n')
print(f'Собственные векторы:\n{v}')
```

```python
def check_for_eig(matr, vect):
    m = np.dot(matr, vect)
    print(f'Линейный оператор A:\n{matr}\n')
    print(f'Вектор X:\n{vect}\n')
    print(f'Произведение A*X:\n{m}\n')
    for i in range(len(vect)):
        if i == 0:
            l = m[i] / vect[i]
        elif l != m[i] / vect[i]:
            print(f'Вектор X={vect} не является собственным вектором линейного оператора A, т.к. {m[i]}/{vect[i]} != {l:.0f}\n')
            return
    print(f'Вектор X={vect} является собственным вектором линейного оператора A, т.к. (A*X)/X={l:.0f}\n')
```

__2.__ Дан оператор поворота на 180 градусов, задаваемый матрицей:

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Покажите, что __любой__ вектор считается для него собственным.

```python
print('Нетрудно заметить, что результатом умножения оператора поворота на вектор (m, n) будет вектор (-m, -n), т.е. lambda = -1.')
print('Покажем это на всех возможных сочетаниях знаков m и n, где m и n не равны 0\n')

a = np.array([[-1, 0], [0, -1]])
x = []
x.append([3, -13])
x.append([23, -31])
x.append([23, 31])
x.append([-23, -31])

for i in range(len(x)):
    print(f'{i+1}.')
    check_for_eig(a, x[i])
```
