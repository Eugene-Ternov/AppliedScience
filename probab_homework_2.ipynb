```python
import math as mt

def combinations(n, k):
    max_val = max(k, n - k)
    min_val = min(k, n - k)
    res = 1
    for i in range(max_val + 1, n + 1):
        res *= i
    return int(res / mt.factorial(min_val))

def print_combinations(n, k):
    return f"{n}!/({k}!*({n}-{k})!)"

def Bernulli(n, k, p):
    return combinations(n, k) * (p ** k) * ((1 - p) ** (n - k))

def Bernulli_ext(n, k, p):
    print(f"\nОтвет. Применим формулу Бернулли для n = {n}, k = {k}, p = {p}, q = {1 - p:.1f}.")
    return Bernulli(n, k, p)

print("1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз.\n   Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.")
n_val = 100
k_val = 85
p = 0.8
res = Bernulli_ext(n_val, k_val, p)
print(f"Вероятность {k_val} попаданий в цель составит {res:.6f}")
```

```python
def Puasson(n, m, p):
    lmbd = n * p
    return (lmbd ** m) * mt.exp(-lmbd) / mt.factorial(m)

print("2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004.\n   В жилом комплексе после ремонта в один день включили 5000 новых лампочек.\n   Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?")
n = 5000
p = 0.0004
m = [0, 2]
print(f"\nОтвет. Применим формулу Пуассона для p = {p}, n = {n}, m = {m[0]} или {m[1]} соответственно.")
print(f"Вероятность того, что не перегорит ни одна лампочка равна {Puasson(n, m[0], p):.6f}.")
print(f"Вероятность того, что перегорят {m[1]} лампочки равна {Puasson(n, m[1], p):.6f}.")
```

```python
print("3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?")
n_val = 144
k_val = 70
p = 0.5
res = Bernulli_ext(n_val, k_val, p)
print(f"Вероятность {k_val} выпаданий орла составит {res:.6f}")
```

```python
print("4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых.\n   Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые?\n   Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?")
n1 = 10
k1 = 7
n2 = 11
k2 = 9
total = combinations(n1, 2) * combinations(n2, 2) 
print(f"\nОтвет. Общее число равновозможных исходов равно {print_combinations(n1, 2)} * {print_combinations(n2, 2)} = {total}.")
white_4 = combinations(k1, 2) * combinations(k2, 2)
print(f"Число исходов, благоприятствующих извлечению 4 белых мячей, равно {print_combinations(k1, 2)} * {print_combinations(k2, 2)} = {white_4}.")
print(f"Вероятность извлечения 4 белых мячей равна {white_4}/{total} = {white_4/total:.6f}.")
print("Далее, 2 белых мяча могут быть извлечены как 2 мяча из 1-го ящика при 2 цветных мячах из 2-го ящика либо 2 мяча из 2-го ящика\nпри 2 цветных мячах из 1-го ящика либо по 1 белому и 1 цветному мячу из каждого ящика.")
white_2 = combinations(k1, 2) * combinations(n2-k2, 2) + (combinations(k2, 2) * combinations(n1-k1, 2)) + (combinations(k1, 1) * combinations(n1-k1, 1) * combinations(k2, 1) * combinations(n2-k2, 1))
print(f"Число исходов, благоприятствующих извлечению 2 белых мячей, равно ({print_combinations(k1, 2)} * {print_combinations(n2-k2, 2)}) +")
print(f"+ ({print_combinations(k2, 2)} * {print_combinations(n1-k1, 2)}) + ({print_combinations(k1, 1)} * {print_combinations(n1-k1, 1)} * {print_combinations(k2, 1)} * {print_combinations(n2-k2, 1)}) = {white_2}.")
print(f"Вероятность извлечения 2 белых мячей равна {white_2}/{total} = {white_2/total:.6f}.")
print(f"Хотя бы один белый мяч будет извлечён всегда за исключением случая, когда из обоих ящиков будет извлечено по 2 цветных мяча.")
white_0 = combinations(n1-k1, 2) * combinations(n2-k2, 2)
print(f"Число исходов, благоприятствующих извлечению 4 цветных мячей, равно {print_combinations(n1-k1, 2)} * {print_combinations(n2-k2, 2)} = {white_0}.")
print(f"Вероятность извлечения хотя бы 1 белого мяча равна 1 - {white_0}/{total} = {1 - white_0/total:.6f}.")
```
